```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/hp/Downloads/archive (6)/PMC Hospital Infrastructure.csv")
df
```

```python
df.select_dtypes(include=object).describe()
```

```python
df.describe()
```

```python
avg_beds_by_types = df.pivot_table(
    values="Number of Beds in facility type",
    index="Type  (Hospital / Nursing Home / Lab)",
    aggfunc="mean"
).round(0)
avg_beds_by_types.sort_values(by="Number of Beds in facility type", ascending=False, inplace=True)
avg_beds_by_types
```

```python
top5=avg_beds_by_types.sort_values(by="Number of Beds in facility type", ascending=False).head(5)
plt.figure(figsize=(20,10))
top5.plot(kind='bar', legend=False,color='green')
plt.title('Top 5 Facility Types by Average Number of Beds', fontsize=16)
plt.xlabel('Facility Type', fontsize=14)
plt.ylabel('Average Number of Beds', fontsize=14)
plt.xticks(rotation=45,ha='right')
plt.grid(axis='y')
plt.show()
```

```python

df["Average Monthly Patient Footfall"] = (
    df["Average Monthly Patient Footfall"]
    .replace(["N.A.", "NA", "n.a.", "N/A"], np.nan)   # handle text "N.A."
)

df["Average Monthly Patient Footfall"] = pd.to_numeric(
    df["Average Monthly Patient Footfall"], errors="coerce"
)


avg_footfall_by_type = df.pivot_table(
    values="Average Monthly Patient Footfall",
    index="Type  (Hospital / Nursing Home / Lab)",
    aggfunc="mean"
).round(0)

print(avg_footfall_by_type)


```

```python
top5_footfall = avg_footfall_by_type.sort_values(by="Average Monthly Patient Footfall", ascending=False).head(5)
plt.figure(figsize=(20,10))
top5_footfall.plot(kind='bar', legend=False, color='orange')
plt.title('Top 5 Facility Types by Average Monthly Patient Footfall', fontsize=16)
plt.xlabel('Facility Type', fontsize=14)
plt.ylabel('Average Monthly Patient Footfall', fontsize=14)
plt.xticks(rotation=45,ha='right')
plt.grid(axis='y')
```
